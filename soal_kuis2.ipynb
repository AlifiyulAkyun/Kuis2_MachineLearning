```python
import pandas as pd
import numpy as np

df = pd.read_excel('data/dataset_quiz.xlsx')

df.head(100)
```

```python
# Cek semua data yang kosong
df.isna().sum() 
```

```python
# Menghilangkan data null
df.drop('id_anggota5', axis=1, inplace=True)

modus_ketua = df['id_ketua'].mode()[0]
df['id_ketua'] = df['id_ketua'].fillna(modus_ketua)

modus_anggota1 = df['id_anggota1'].mode()[0]
df['id_anggota1'] = df['id_anggota1'].fillna(modus_anggota1)

modus_anggota2 = df['id_anggota2'].mode()[0]
df['id_anggota2'] = df['id_anggota2'].fillna(modus_anggota2)

modus_anggota3 = df['id_anggota3'].mode()[0]
df['id_anggota3'] = df['id_anggota3'].fillna(modus_anggota3)

modus_anggota4 = df['id_anggota4'].mode()[0]
df['id_anggota4'] = df['id_anggota4'].fillna(modus_anggota4)
# Cek Data
df.head()
```

```python
# Cek seluruh data yang kosong
df.isna().sum()
```

## Soal 1 & 2<br>
Pemilahan data training dan testing yang sesuai dan proporsional, Menentukan validasi data

```python
# Split data
from sklearn.model_selection import train_test_split

# Split data training dan dan lainnya
# data lainnya di split lagi menjadi validasi dan testing.
# Rasio yang digunakan adalah 8:1:1
df_train, df_unseen = train_test_split(df, test_size=0.2, random_state=0)

# Split lagi antara validasi dan testing
df_val, df_test = train_test_split(df_unseen, test_size=0.5, random_state=0)

# Cek masing-masing ukuran data
print(f'Jumlah dari data asli: {df.shape[0]}')
print(f'Jumlah dari data training: {df_train.shape[0]}')
print(f'Jumlah dari data testing: {df_test.shape[0]}')
print(f'Jumlah dari data validasi: {df_val.shape[0]}')

# Cek rasio tiap label
print('=========')
print(f'Jumlah dari label data asli:\n{df.keterlibatan_vltr.value_counts()}')
print(f'Jumlah dari label data training:\n{df_train.keterlibatan_vltr.value_counts()}')
print(f'Jumlah dari label data testing:\n{df_test.keterlibatan_vltr.value_counts()}')
print(f'Jumlah dari label data validasi:\n{df_val.keterlibatan_vltr.value_counts()}')
```

## Soal 3 <br>
Menentukan kolom sehingga menjadi "label"

```python
from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()

df['keterlibatan_vltr'] = labelencoder.fit_transform(df['keterlibatan_vltr'])
df['skema'] =labelencoder.fit_transform(df['skema'])
df['luaran_hki'] = labelencoder.fit_transform(df['luaran_hki'])
df['jenis_luaran_hki'] = labelencoder.fit_transform(df['jenis_luaran_hki'])
df['potensi_hki'] = labelencoder.fit_transform(df['potensi_hki'])
df['luaran_lain'] = labelencoder.fit_transform(df['luaran_lain'])
df['jenis_luaran_lain'] = labelencoder.fit_transform(df['jenis_luaran_lain'])
df['status_luaran_lain'] = labelencoder.fit_transform(df['status_luaran_lain'])
df.head()
```

## Soal 4<br>
Menerapkan dua jenis metode klasifikasi dan lakukan perbandingan akurasi dengan wajib melibatkan variabel keterlibatan volunteer sampai variabel luaran-luaran yang dihasilkan

```python
#selesksi fitur
X = df.iloc[:, 6:]
y = df['keterlibatan_vltr']

X.shape
```

```python
#training dan validasi data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.5, random_state=4)

X_val, X_unseen, y_val, y_unseen = train_test_split(X_test, y_test, test_size=0.6, random_state=5)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

# Sesuaikan dt ke set training
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)

# Memprediksi label set test
y_pred_dt = dt.predict(X_test)

# Menghitung set accuracy
acc_dt = accuracy_score(y_test, y_pred_dt)
print("Test set accuracy: {:.2f}".format(acc_dt))
print(f"Test set accuracy: {acc_dt}")
```

```python
from sklearn.ensemble import RandomForestClassifier

# Sesuaikan rf ke set training
rf = RandomForestClassifier(n_estimators=30, random_state=12)
rf.fit(X_train, y_train)

# Memprediksi label set test
y_pred_rf = rf.predict(X_test)

# Menghitung set accuracy
acc_rf = accuracy_score(y_test, y_pred_rf)
print("Test set accuracy: {:.2f}".format(acc_rf))
print(f"Test set accuracy: {acc_rf}")
```

## Soal 5
Melakukan tunning hyperparameter berdasarkan hasil klasifikasi pada nomor 3 menggunakan metode Grid Search dan Random Search

```python
# GridSearch menggunakan Decission Tree
from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RepeatedKFold

#inisiasi repeated k-fold
cv = RepeatedKFold(n_splits=4, n_repeats=3, random_state=42)

# Definisikan hyperparameter yang akan digunakan
params = {
    'criterion' : ['gini', 'entropy', 'log_loss'],
    'max_depth' : list(range(5,11))
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid = GridSearchCV(dt, param_grid=params, cv=cv)

# Fit / latih berdasarkan grid
%timeit grid.fit(X,y)

# Evaluasi dengan score
score = grid.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {grid.best_params_}')
```

```python
# Random search menggunakan Decission Tree
from sklearn.model_selection import RandomizedSearchCV

# Definisikan hyperparameter yang akan digunakan
params = {
    'criterion' : ['gini', 'entropy'],
    'max_depth' : list(range(5,11))
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
random = RandomizedSearchCV(dt, params, cv=cv)

# Fit / latih berdasarkan random
%timeit random.fit(X,y)

# Evaluasi dengan score
score = random.score(X,y)

print(f'Hasil evaluasi: {score}')
print(f'Konfigurasi hyperparameter : {random.best_params_}')
```

```python
# Grid Search menggunakan Random Forest
# Definisikan hyperparameter yang akan digunakan
params_rf = {
    'n_estimators': [5,20,50,70],
    'max_depth': [int(x) for x in np.linspace(10, 120, num = 12)],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
grid_rf = GridSearchCV(rf, param_grid=params_rf)

# Fit / latih berdasarkan grid
%timeit grid_rf.fit(X,y)

# Evaluasi dengan score
score_rf = grid_rf.score(X,y)

print(f'Hasil evaluasi: {score_rf}')
print(f'Konfigurasi hyperparameter : {grid_rf.best_params_}')
```

```python
# Random Search menggunakan Random Forest
# Definisikan hyperparameter yang akan digunakan
params_rf = {
    'n_estimators': [5,20,50,70],
    'max_depth': [int(x) for x in np.linspace(10, 120, num = 12)],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    'bootstrap': [True, False]
}

# inisiasi grid berdasarkan nilai repeated k-fold dan hyperparameter
random_rf = RandomizedSearchCV(rf, params_rf)

# Fit / latih berdasarkan grid
%timeit random_rf.fit(X,y)

# Evaluasi dengan score
score_rf = random_rf.score(X,y)

print(f'Hasil evaluasi: {score_rf}')
print(f'Konfigurasi hyperparameter : {random_rf.best_params_}')
```
